SciPy :

Le test de corrlation de Pearson entre la surface et le prix a donn un coefficient de corrlation trs lev (0.996), indiquant une forte corrlation positive. La p-value trs faible (< 0.001) suggre que cette corrlation est statistiquement significative.

Le test de normalit de Shapiro-Wilk sur la distribution des pr ix a donn une statistique de 0.955 et une p-value de 0.00185, ce qui signifie que les prix ne suivent pas exactement une distribution normale (selon un seuil de significativit de 0.05)

```python
from scipy.stats import pearsonr

# Calculer le coefficient de corrlation de Pearson entre surface et prix
corr, p_value = pearsonr(df['surface'], df['price'])

# Affichage des rsultats
print(f"Coefficient de corrlation de Pearson: {corr}, p-value: {p_value}")
```

 Test t de Student (t-test)
Le test t de Student est utilis pour comparer les moyennes de deux chantillons et voir s'il existe une diffrence statistiquement significative entre ces deux groupes.

```python
from scipy.stats import ttest_ind

# Diviser les appartements en deux groupes selon le nombre de chambres
group1 = df[df['nbRooms'] > 3]['price']
group2 = df[df['nbRooms'] <= 3]['price']

# Effectuer le test t
stat, p_value = ttest_ind(group1, group2)

# Rsultat
print(f"Statistique du test t: {stat}, p-value: {p_value}")
```

Ce test compare les moyennes de deux groupes qui sont lis, par exemple avant et aprs une intervention.

Exemple d'utilisation : Si tu avais les prix d'appartements avant et aprs une rnovation, tu pourrais utiliser ce test pour voir si la rnovation a un impact significatif sur les prix.
